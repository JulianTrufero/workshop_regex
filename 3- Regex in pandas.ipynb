#### Replace in pandas

How to replace a string in pandas?

In pandas we have the `str.replace` method that allows us to repalce each ocurrence of pattern/regex in the Series 
- Replace character/s under a single DataFrame column:


```python
df['column name'] = df['column name'].str.replace('old character','new character', regex = True)
```
- Replace character/s under the entire DataFrame:

```python
df = df.replace('old character','new character', regex=True)
```



More info [here](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html)
